```python
###Position map of evolutionary tracks###

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import sys
sys.path.append("../research2017/ELVIS_HALO_Catalogs/IsolatedCatalogs")
sys.path.append("../research2017/ELVIS_Main_Branches/IsolatedTrees")
```

```python
def posTimeArray(system,scFactor):
    #Create array of X, Y, Z coordinates
    dataX=np.genfromtxt("../ELVIS_Main_Branches/IsolatedTrees/"+system+"/X.txt");#get data of X position
    dataY=np.genfromtxt("../ELVIS_Main_Branches/IsolatedTrees/"+system+"/Y.txt");#get data of Y position
    dataZ=np.genfromtxt("../ELVIS_Main_Branches/IsolatedTrees/"+system+"/Z.txt");#get data of Z position
    dataScale=np.genfromtxt("../ELVIS_Main_Branches/IsolatedTrees/"+system+"/scale.txt"); #get data of scale factor
    selec=(dataScale[0]==scFactor)
    index=np.where(selec)
    n=index[0][0]
    position=np.vstack((dataX[:,n],dataY[:,n],dataZ[:,n]))
    return position
```

```python
def plotPosTime(system,scFactor):
    #Plot map of halos as given scale factor
    a0=posTimeArray(system,scFactor);
    fig=plt.figure();
    ax=fig.add_subplot(111,projection='3d');
    plt.title('Position of resolved halos at scale factor'+' '+str(scFactor));
    ax.set_xlim(30,40)
    ax.set_ylim(30,40)
    ax.set_zlim(30,40)
    ax.set_xlabel('X (Mpc)');
    ax.set_ylabel('Y (Mpc)');
    ax.set_zlabel('Z (Mpc)');
    ax.scatter(a0[0],a[1],a[2],'.',color='b')
    plt.show()
```

```python
plotPosTime('iLincoln',0.40570)
```

```python
def calcEnvelope(system,scFactor):
    #Calculate distances of outermost halos. (Xmax-Xmin, Ymax-Ymin, Zmax-Zmin)
    arr=posTimeArray(system,scFactor);
    conditionX=(arr[0]>0);
    conditionY=(arr[1]>0);
    conditionZ=(arr[2]>0);
    xClip=np.extract(conditionX,arr[0]);
    yClip=np.extract(conditionY,arr[1]);
    zClip=np.extract(conditionZ,arr[2]);
    xMin=np.amin(xClip);
    yMin=np.amin(yClip);
    zMin=np.amin(zClip);
    xMax=np.amax(xClip);
    yMax=np.amax(yClip);
    zMax=np.amax(zClip);
    xt=xMax-xMin;
    yt=yMax-yMin;
    zt=zMax-zMin;
    envelope=[[xMin,xMax],[yMin,yMax],[zMin,zMax]];
    tot=(xt,yt,zt)
    return envelope, tot
```

```python
k=calcEnvelope('iLincoln',0.40570)
```
